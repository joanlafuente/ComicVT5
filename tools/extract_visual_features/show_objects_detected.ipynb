```python
import matplotlib.pyplot as plt
import h5py
import numpy as np
from PIL import Image
```

```python
h5f = h5py.File("/home/jlafuente/ComicsQA/Datasets/comics/frcnn_features/boxes10.h5", "r")

for i, k in enumerate(h5f.keys()):
    if i == 600:
        key = k
        boxes = h5f[f"{k}/boxes"][:]
        features = h5f[f"{k}/features"][:]
        w = h5f[f"{k}/img_w"][()]
        h = h5f[f"{k}/img_h"][()]
        break
h5f.close()

boxes[:, (0, 2)] /= w
boxes[:, (1, 3)] /= h
book_id, page_id, panel_id = key.split("_")
img = Image.open(f"/home/jlafuente/ComicsQA/Datasets/comics/panels/{book_id}/{page_id}_{panel_id}.jpg")
print(img.size)
boxes[:, (0, 2)] *= img.size[0]
boxes[:, (1, 3)] *= img.size[1]
plt.imshow(img)
plt.axis("off")
plt.title(f"Book: {book_id}, Page: {page_id}, Panel: {panel_id}")
plt.show()

fig = plt.figure(figsize=(15, 8))
plt.title("Objects detected")
plt.axis("off")
for i, box in enumerate(boxes):
    ax = fig.add_subplot(2, int(len(boxes)/2), i+1)
    ax.imshow(img.crop(box))
    ax.axis("off")
plt.show()
```
